# <a name="overview"> Описание проекта</a>

В интернет-магазине "Стримчик" идут продажи компьютерных игр по всему миру. 


Из открытых источников доступны исторические данные о продажах игр, оценки пользователей и экспертов, жанры и платформы (например, Xbox или PlayStation). 


Необходимо выявить выявить определяющие успешность игры закономерности. Это позволит сделать ставку на потенциально популярный продукт и спланировать рекламные кампании.

Для проведения анализа используются данные до 2016 года. Представим, что сейчас декабрь 2016 г., и планируется кампания на 2017-й.

В наборе данных попадается аббревиатура ESRB (Entertainment Software Rating Board) — это ассоциация, определяющая возрастной рейтинг компьютерных игр. ESRB оценивает игровой контент и присваивает ему подходящую возрастную категорию, например, «Для взрослых», «Для детей младшего возраста» или «Для подростков».

## Описание данных

* `Name` — название игры
* `Platform` — платформа
* `Year_of_Release` — год выпуска
* `Genre` — жанр игры
* `NA_sales` — продажи в Северной Америке (миллионы проданных копий)
* `EU_sales` — продажи в Европе (миллионы проданных копий)
* `JP_sales` — продажи в Японии (миллионы проданных копий)
* `Other_sales` — продажи в других странах (миллионы проданных копий)
* `Critic_Score` — оценка критиков (максимум 100)
* `User_Score` — оценка пользователей (максимум 10)
* `Rating` — рейтинг от организации ESRB (англ. Entertainment Software Rating Board). Эта ассоциация определяет рейтинг компьютерных игр и присваивает им подходящую возрастную категорию.


Данные за 2016 год могут быть неполными.

## <a name="res-01-pl">Продажи по платформам</a> 

Посмотрим на суммарные продажи по платформам в год. Сначала посмотрю на наибольшие суммарные продажи по платформам.

```python
# сгруппируем данные количеству продаж для каждой платформы
total_sales_pl = (games_df.pivot_table(index = 'platform', aggfunc = {'total_sales':'sum'})
                   .rename(columns = {'total_sales':'sum'})                
                   .sort_values(by = 'sum'))
```

Самыми популярными платформами являются: `PS`, `DS`, `Wii`, `PS3`, `X360`, `PS2`. Показатели продаж именно для этих платформ резко идут вверх на графике. Именно для них я проверю расределение по годам. 

```python
# выбираю все популярные платформы
popular_platforms = ['PS', 'DS', 'Wii', 'PS3', 'X360', 'PS2']

#собираю в отдельный датасет 
popular_platforms_releases = games_df.query('platform in @popular_platforms')
```

```python
# собираю всё в сводную таблицу и строю график
ppr_plot = popular_platforms_releases.pivot_table(index='year_of_release', columns = 'platform', aggfunc = {'total_sales':'sum'})
ppr_plot.plot(figsize = (20,6), xlim = (1993, 2016)).set_title('Суммарные продажи для TOP-платформ')
plt.xlabel('Год выпуска', fontsize=15)
plt.ylabel('Суммарное количество продаж', fontsize=15)
plt.legend()
plt.show()
```

# <a name="hyp">Проверка гипотез</a>   

##  <a name="hyp-rating">Средние пользовательские рейтинги платформ Xbox One и PC одинаковые</a>    

Средние пользовательские рейтинги платформ Xbox One и PC одинаковые:

* **H0** Средние пользовательские рейтинги пользователей платформ Xbox One & PC равны
* **H1** Средние пользовательские рейтинги пользователей платформ Xbox One & PC не равны

Сравним дисперсии выборок, чтобы выбрать значение аргумента `equal_var`.

```python
hyp_pl = ['PC', 'XOne']
# таблица с данными о среднем, медианном, стандартном отклонении и дисперсии
ppr_ayear.query('platform in @hyp_pl').groupby('platform')\
     .agg({'user_score':['var','std']})\
     .style.format("{:.1f}")
```

Различие значительное, значит  `equal_var = False`

Так как две выборки независимы, использую Student t-test.

Сначала проверим данные о выручке на выбросы. С моей точки зрения, лучше всего для этого подходит boxplot.

```python
# создаю отдельные датасеты по платформам
xone = ppr_ayear.query('platform == "XOne"').dropna(subset=['user_score'])
pc = ppr_ayear.query('platform == "PC"').dropna(subset=['user_score'])
                     
# смотрю на ящики с усами
fig, axs = plt.subplots(1,2, figsize=(10,5))
xone.boxplot('user_score', by='platform', ax = axs[0])
pc.boxplot('user_score', by='platform', ax = axs[1])
plt.show()                     
    
```

<div class="alert alert-block alert-success">
<h2>Комментарий от ревьювера (все здОрово)</h2> 

О, субплоты - супер)

На графиках видно, что есть выбросы. Попробуем от них избавиться.

```python
# функция для поиска выбросов 
def remove_outlier(df, column, graf=0, x = 10, y = 10):
    # присваиваю переменным значение квартилей
    low_quartile = df[column].quantile(0.25)
    upper_quartile = df[column].quantile(0.75)
    inter_q_range = upper_quartile - low_quartile
    # использую Interquartile Rule to Find Outliers: https://www.thoughtco.com/what-is-the-interquartile-range-rule-3126244
    find_outlier = (df[column] >= (low_quartile - 1.5 * inter_q_range)) & (df[column] <= (upper_quartile + 1.5 * inter_q_range))
    if graf == 1:
        return df.loc[find_outlier].boxplot([column], figsize=(x,y))
    else:
        return df.loc[find_outlier]
```

```python
# удаляем выбросы
xone = remove_outlier(xone, 'user_score', 0)
xone_score = xone['user_score']


pc = remove_outlier(pc, 'user_score', 0)
pc_score = pc['user_score']
```

```python
# проверяем результат
fig, axs = plt.subplots(1,2, figsize=(10,5))
xone.boxplot('user_score', by='platform', ax = axs[0])
pc.boxplot('user_score', by='platform', ax = axs[1])
plt.show()  
```

Вроде выбросов меньше, можно приступать к проверке гипотезы.

```python
#проверяем гипотезу
alpha = 0.05
result = st.ttest_ind(xone_score, pc_score, equal_var=False)

if result.pvalue < alpha:
    print('p-value:', result.pvalue)
    print('Отвергаем H0')
else:
    print('p-value:', result.pvalue)
    print('Отвергаем H1')
```

Значимость p-value больше уровня значимости, значит отвергнуть нулевую гипотезу мы не можем

Вывод: пользовательские рейтинги для платформ XOne и PC не различаются.
